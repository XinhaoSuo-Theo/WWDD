## 2.1 定义GINI系数

```python
def gini_mat(x,axis=2):
    '''
    本函数用于计算数组形式的基尼系数
    x为输入数据（np.array）
    axis为数据所在维度(哪些数字应该累计起来)
    '''  
    n=np.sum(x>0,axis)##湿日天数
    sorted_x = np.sort(x,axis)
    cumx = np.cumsum(sorted_x,axis, dtype=float)
    xmax=np.nanmax(cumx,axis)
    # 均等权重
    if len(x.shape) == axis+1:
        results = (n + 1 - 2 * np.nansum(cumx,axis) / xmax) / n #目前只支持数据放在最后一维
    else:
        print('数组维度顺序有误')
        results = -1

    return results
```

```python
time_start = time.time()
gini_wet=np.zeros((70,181,360))
gini_dry=np.zeros((70,181,360))
for year in range(1951,2021):#2020\
    print(year)
    wet_day=np.zeros((181,360))
    dry_day=np.zeros((181,360))
    
    n=year-1949
    nyear=year-1951
    if year%4==0:
        run_year=n//4
        start=nyear*365+run_year*1
        data_year=pre_end[:,:,start:start+366]
    else:
        run_year=n//4
        start=nyear*365+run_year*1
        data_year=pre_end[:,:,start:start+365]
    data_pre0=np.where(data_year!=0,data_year,np.nan)        
    wet_day=np.sum(data_year>0,axis=2)##湿日天数
    dry_day=np.sum(data_year==0,axis=2)### 计算降雨日为0的天数
    gini_wet[year-1951,:,:]=gini_mat(data_pre0)## 降雨日的基尼系数
    gini_dry[year-1951,:,:]=dry_day[:,:]/(dry_day[:,:]+wet_day[:,:])
    
    
    
time_end = time.time()  
time_c= time_end - time_start   #运行所花时间
print('time cost', time_c, 's')
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# 定义颜色和对应的值
colors = [ '#d2d2d2','#8cc2d7','#a2dfa8','#f06948']  # 对应的颜色
values = [1,2,3,4]  # 数据值

# 创建离散 colormap
discrete_cmap = mcolors.ListedColormap(colors, name="discrete_cmap")

# 定义边界（每个值的上下边界，确保覆盖所有值）
boundaries = [0.5, 1.5, 2.5, 3.5, 4.5]  # 值之间的边界
norm = mcolors.BoundaryNorm(boundaries, len(colors))

# 创建示例数据（随机生成包含这些值的矩阵）
data = np.random.choice(values, size=(10, 10))  # 替换为你的实际地图数据

# 绘制图像，应用自定义 colormap 和边界
plt.figure(figsize=(6, 5))
plt.imshow(data, cmap=discrete_cmap, norm=norm)
cbar = plt.colorbar(ticks=values)  # 设置颜色条刻度
cbar.ax.set_yticklabels([str(v) for v in values])  # 设置颜色条标签
plt.title('Map with Discrete Values (4, 3, 2, 1)')
plt.show()
```
